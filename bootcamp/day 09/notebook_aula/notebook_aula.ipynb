```python
formato_longo = pd.DataFrame([['ce', 'fortaleza', 'casos', 10.0],
                              ['rj', 'rio de janeiro', '칩bitos', 2.0],
                              ['df', 'bras칤lia', '칩bitos', 3.0],
                              ['rn', 'natal', 
                               'habitantes vacinados', 300.0]],
                             columns=['estado', 'municipio', 
                                      'vari치vel', 'valor'])
```

```python
formato_longo
```

```python
formato_largo = formato_longo.set_index(['estado', 'municipio', 'vari치vel']).unstack()
formato_largo
```

```python
#comando "tradicional" de transformar formato largo em formato longo
formato_largo.droplevel(level=0, axis='columns').reset_index().melt(id_vars=['estado', 'municipio'], var_name='vari치vel', value_name='valor').dropna()
```

```python
#equivalente ao comando acima, mas com o truque da Daiana 游뱔
formato_largo.stack()
```
