### Opera√ß√µes b√°sicas sobre os dados

Realize as seguintes opera√ß√µes:
1. Encontre os valores de passagem mais baratos e mais caros pagos pelos passageiros do Titanic. Use as fun√ß√µes **idxmin()** e **idxmax()** para tal.
2. Calcule a m√©dia de idade dos passageiros do Titanic.

Confira sua resposta com a sa√≠da da fun√ß√£o **describe()**.

````python
idx_mais_barato = titanic_df['Fare'].idxmin()
valor_mais_barato = titanic_df.iloc[idx_mais_barato,9]
print(f'O valor mais barato de passagem pago por um passageiro foi {valor_mais_barato}.')

idx_mais_caro = titanic_df['Fare'].idxmax()
valor_mais_caro = titanic_df.iloc[idx_mais_caro,9]
print(f'O valor mais caro de passagem pago por um passageiro foi {valor_mais_caro}.')
````

````python
media_idades = titanic_df['Age'].mean(skipna=True)
print(f'A m√©dia de idade dos passageiros do Titanic era {media_idades}.')
````
